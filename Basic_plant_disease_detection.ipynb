```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("Please upload your leaf image (jpg/png)...")
uploaded = files.upload()

image_path = list(uploaded.keys())[0]

leaf = cv2.imread(image_path)
if leaf is None:
    print("Error: Image not loaded. Check the filename and try again.")
else:
    print("Image loaded successfully!")

    leaf = cv2.cvtColor(leaf, cv2.COLOR_BGR2RGB)

    hsv = cv2.cvtColor(leaf, cv2.COLOR_RGB2HSV)
    green_channel = hsv[:, :, 1]

    avg_green = np.mean(green_channel)

    threshold = 100
    status = "Healthy Leaf ✅" if avg_green > threshold else "Possibly Diseased Leaf ❌"

    plt.imshow(leaf)
    plt.title(f"Leaf Analysis: {status}")
    plt.axis("off")
    plt.show()

    print(f"Average Green Intensity: {avg_green:.2f}")
    print(f"Classification: {status}")

```
